```python
#Converting into snake_case
df.rename(columns = {'normalized-losses':'normalized_losses',
                    'fuel-type':'fuel_type',
                     'num-of-doors':'num_of_doors',
                     'body-style':'body_style',
                     'drive-wheels':'drive_wheels',
                     'engine-location':'drive_wheels',
                     'engine-location':'engine_location',
                     'wheel-base':'wheel_base',
                     'curb-weight':'curb_weight',
                     'engine-type':'engine_type',
                     'fuel-system':'fuel_system',
                     'compression-ratio':'compression_ratio',
                     'peak-rpm':'peak_rpm',
                     'city-mpg':'city_mpg',
                     'highway-mpg':'highway_mpg',
                     'engine-size':'engine_size',
                     'num-of-cylinders':'num_of_cylinders'
                    }, inplace = True)
```

```python
#Replacing NaN value with mean value for normalized loss
df["normalized_losses"] = pd.to_numeric(df["normalized_losses"], errors='coerce')
df["normalized_losses"] = df["normalized_losses"].astype(float)
df["normalized_losses"].fillna(df["normalized_losses"].mean(), inplace=True)
```

```python
#Replacing NaN value with mean value for price
df["price"] = pd.to_numeric(df["price"], errors='coerce')
df["price"] = df["price"].astype(float)
df["price"].fillna(df["price"].mean(), inplace=True)
```

```python
#Replacing NaN value with common four number of door in car
df['num_of_doors'] = df['num_of_doors'].replace(np.nan,'four')
```

```python
##Replacing NaN value with mean value in bore
df["bore"] = pd.to_numeric(df["bore"], errors='coerce')
df["bore"] = df["bore"].astype(float)
df["bore"].fillna(df["bore"].mean(), inplace=True)
```

```python
#Replacing NaN value with mean value in stroke
df["stroke"] = pd.to_numeric(df["stroke"], errors='coerce')
df["stroke"] = df["stroke"].astype(float)
df["stroke"].fillna(df["stroke"].mean(), inplace=True)
```

```python
#Replacing NaN value with mean value in horsepower
df["horsepower"] = pd.to_numeric(df["horsepower"], errors='coerce')
df["horsepower"] = df["horsepower"].astype(float)
df["horsepower"].fillna(df["horsepower"].mean(), inplace=True)
```

```python
#Replacing NaN value with mean value in peak rpm
df["peak_rpm"] = pd.to_numeric(df["peak_rpm"], errors='coerce')
df["peak_rpm"] = df["peak_rpm"].astype(float)
df["peak_rpm"].fillna(df["peak_rpm"].mean(), inplace=True)
```
